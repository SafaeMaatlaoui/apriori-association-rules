```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv('bread basket.csv')
print(data.shape)
data.head(10)
```

```python
print('Le nombre total d\'éléments vendus à la boulangerie est:',data['Item'].nunique())
```

```python
print('Dix éléments les plus vendus à la boulangerie')
data['Item'].value_counts().head(10)
```

```python
#2.2 Transformation en tableau binaire

# Define dataset to machine learning
data_p = data.groupby(["Transaction","Item"]).size().reset_index(name="Count")

Transaction_Item = (data_p.groupby(['Transaction', 'Item'])['Count'].sum().unstack().reset_index().fillna(0).set_index('Transaction'))
```

```python
def encode_data(datapoint):
    if datapoint <= 0:
        return 0
    else:
        return 1

TC = Transaction_Item.applymap(encode_data)

TC.iloc[:15,:]
```

```python
print("SHAPE : ",TC.shape)
```

```python
#colonnes de TC 
print(TC.columns)
```

```python
#3.1 Extraction des itemsets

#importation de la fonction apriori
from mlxtend.frequent_patterns import apriori

#itemsets frequents

freq_itemsets = apriori(TC,min_support=0.01, use_colnames=True) 
print(type(freq_itemsets))
```

```python
print(freq_itemsets.columns)
print(freq_itemsets.shape)
```

```python
#affichage des 15 premiers itemsets
freq_itemsets.head(14)
```

```python
#3.2 Filtrage des itemsets
#type du champ 'itemsets'
print(type(freq_itemsets.itemsets))
```

```python
#accès indexé au premier élément
print(freq_itemsets.itemsets[0])
```

```python
#3.2.1  Solution 1 – Utilisation de la méthode apply() de pandas.Series
#fonction de test d'inclusion
def is_inclus(x,items):
    return items.issubset(x)
```

```python
#recherche des index des itemsets correspondant à une condition
id = np.where(freq_itemsets.itemsets.apply(is_inclus,items={"Cookies"}))
print(id)
```

```python
#affichage des itemsets corresp.
freq_itemsets.loc[id]
```

```python
#3.2.2 Solution 1 – Utilisation de la fonction anonyme lambda

#passer par une fonction lambda si on est pressé
np.where(freq_itemsets.itemsets.apply(lambda x,ensemble:ensemble.issubset(x),ensemble={"Cookies"}))
```

```python
#3.2.3 Solution 2 – Utilisation des opérateurs de comparaison de pandas.Series

#itemsets contenant la ram 8 - passer par les méthodes natives de Series
freq_itemsets[freq_itemsets['itemsets'].ge({"Cookies"})]
```

```python
#itemset avec la ram 8
freq_itemsets[freq_itemsets['itemsets'].eq({"Cookies"})]
```

```python
#itemsets contenant la ram 4 et 8
freq_itemsets[freq_itemsets['itemsets'].ge({"Cookies","Coffee"})]
```

```python
#4.1 Génération des règles

#fonction de calcul des règles
from mlxtend.frequent_patterns import association_rules
#génération des règles à partir des itemsets fréquents
regles = association_rules(freq_itemsets,metric="lift", min_threshold=1)
```

```python
#type de l'objet renvoyé
print(type(regles))
```

```python
#dimension
print(regles.shape)
```

```python
#liste des colonnes
print(regles.columns)
```

```python
#5 "premières" règles
regles.iloc[:5,:]
```

```python
#règles en restreignant l'affichage à qqs colonnes
myRegles = regles.loc[:,['antecedents','consequents','lift']]
print(myRegles.shape)
```

```python
#pour afficher toutes les colonnes
pd.set_option('display.max_columns',5)
pd.set_option('precision',3)

#affichage des 5 premières règles
myRegles[:5]
```

```python
#4.2 Filtrage des règles

#affichage des règles avec un LIFT supérieur ou égal à 2
print(myRegles[myRegles['lift'].ge(1.9)])
#La combinaison entre Tea,Coffee et Cake a un lift maximale de 1,938, ce qui signifie que dans la plupart des cas, 
#les personnes qui achètent le (Coffee,Tea) achètent également du (Cake)
#Lift 1,938 signifie que le (Coffee,Tea) a 1,938 fois plus de chances d'être apporté lorsque le (Cake) est apporté.
```

```python
#afficher les règles de conséquent {8}
print(myRegles[myRegles['consequents'].eq({"Cake"})])
```

```python
#filtrer les règles contenant '8' dans leur antécédent
print(myRegles[myRegles['antecedents'].ge({"Cake"})])
```
